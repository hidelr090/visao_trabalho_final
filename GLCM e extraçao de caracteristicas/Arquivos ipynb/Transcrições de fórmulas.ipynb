```python
import numpy as np
```

```python
image = np.array([
            [0.07,0.14,0,0],
            [0.14,0.36,0,0],
            [0,0,0.21,0],
            [0,0,0,0.07],
])

print(image)

#b = (image - np.min(image))/np.ptp(image)
#print(b)
```


<center>Uniformidade</center>

```python
def uniformity(image):
    sum_index_values = 0
    for i in image:
        for j in i:
            sum_index_values += j**2
    return sum_index_values

print(uniformity(image))
```


<center>Entropy</center>

```python
def entropy(image):
    sum_index_values = 0
    for i in image:
        for j in i:
            if j > 0:
                sum_index_values += j*np.log10(j)
    return -(sum_index_values)

print(entropy(image))
```


<center>Correlação</center>

```python
def correlation(image):
    sum_index_values = 0

    for i in range(len(image)):
        for j in range(len(image[i])):
            if image[i,j] != 0:
                sum_index_values += (i - np.mean(image[i])) * (j - np.mean(image[:,j])) * image[i,j]/(np.std(image[i])*np.std(image[:,j]))
    return sum_index_values
    
correlation(image)
```


<center>Contraste</center>

```python
def contrast(image):
    sum_index_values = 0
    
    for i in range(len(image)):
        for j in range(len(image[i])):
            sum_index_values += ((i - j) ** 2) * image[i,j]
    return sum_index_values

contrast(image)
```
